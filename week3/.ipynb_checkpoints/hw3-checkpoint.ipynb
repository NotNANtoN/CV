```python
from scipy.ndimage import gaussian_filter
def gaussian_pyramid(img, num_levels):
    imgs = []
    for _ in range(num_levels):
        print(img.shape)
        smoothed = gaussian_filter(img, 1)
        imgs.append(smoothed)
        img = img[::2, ::2]
        
    for img in imgs[::-1]:
        plt.imshow(img, cmap="gray")
        plt.show()
        
```

```python
gaussian_pyramid(I, 4)
```

```python
from scipy.ndimage import gaussian_filter
def laplacian_pyramid(img, num_levels):
    imgs = []
    for _ in range(num_levels):
        print(img.shape)
        smoothed1 = gaussian_filter(img, 1).astype(float)
        smoothed2 = gaussian_filter(img, 2).astype(float)
        dog = img - smoothed1
        

        imgs.append(dog)
        img = img[::2, ::2]
        
    for img in imgs[::-1]:
        plt.imshow(img, cmap="gray")
        plt.show()
        
```

```python
laplacian_pyramid(I, 4)
```
