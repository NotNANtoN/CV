```python
import imageio
import numpy as np
import matplotlib.pyplot as plt
```

```python
woman = imageio.imread("woman.tif")
saturn = imageio.imread("saturn.tif")
mt = imageio.imread("MaruTaro.jpg").mean(-1)
```

```python
woman.shape
```

```python
woman_fft = np.fft.fft2(woman)
```

```python
woman_freq = np.abs(woman_fft)
```

```python
woman_phase = np.angle(woman_fft)
```

```python
plt.imshow(np.log(woman_freq), cmap="gray")
```

```python
plt.imshow(woman_phase, cmap="gray")
```

```python
def Fuv(freq, phase):
    return freq * np.exp(1j * phase)
```

```python
woman_fft_messy = Fuv(woman_freq, woman_phase)
```

```python
woman_ifft = np.fft.ifft2(woman_fft_messy)
```

```python
plt.imshow(woman_ifft.real, cmap="gray")
```

```python
plt.imshow(np.fft.ifft2(Fuv(1, woman_phase)).real, cmap="gray")
```

```python
plt.imshow(np.fft.ifft2(np.fft.ifftshift(Fuv(np.angle(np.fft.fftshift(woman_fft)), 0))).real, cmap="gray")
```

```python
woman_rot = np.rot90(np.rot90(woman))
```

```python
plt.imshow(np.log(np.abs(np.fft.fft2(woman_rot))).real, cmap="gray")
```

```python
plt.imshow(np.angle(np.fft.fft2(woman_rot)).real, cmap="gray")
```

## 2 Filter Saturn

```python
saturn_fft = np.fft.fft2(saturn)
```

```python
saturn_freq_s = np.fft.fftshift(np.abs(saturn_fft))
```

```python
saturn_phase_s = np.fft.fftshift(np.angle(saturn_fft))
```

```python
plt.figure(figsize=(10,10))
plt.imshow(np.log(saturn_freq_s), cmap="gray")
```

```python
notchie = np.ones_like(saturn)
```

```python
h = notchie.shape[0]
```

```python
w = notchie.shape[1]
```

```python
notchie[:h // 2 - 5, w // 2] = 0
```

```python
notchie[h // 2 + 5:, w // 2] = 0
```

```python
plt.figure(figsize=(10,10))
plt.imshow(notchie, cmap="gray")
```

```python
filtered_saturn = notchie * saturn_freq_s
```

```python
plt.rcParams['figure.figsize'] = (10.0, 10.0)
plt.imshow(np.log(filtered_saturn.real), cmap="gray")
```

```python
real_saturn = np.fft.ifft2(np.fft.ifftshift(Fuv(filtered_saturn, saturn_phase_s)))
```

```python
plt.imshow(saturn.real, cmap="gray")
```

```python
plt.imshow(real_saturn.real, cmap="gray")
```

```python
strer = np.fft.ifft2(np.fft.ifftshift(Fuv(saturn_freq_s * (1 - notchie), saturn_phase_s)))
```

```python
plt.imshow(strer.real, cmap="gray")
```
