```python
import imageio
from skimage.feature import match_template
import matplotlib.pyplot as plt
import numpy as np
```

```python
im = imageio.imread("coco264316.jpg").mean(axis=-1)
clock = imageio.imread("coco264316clock.jpg").mean(axis=-1)
```

```python
plt.imshow(im, cmap="gray")
```

```python
plt.imshow(clock, cmap="gray")
```

```python
matched = match_template(im, clock)
```

```python
plt.imshow(matched, cmap="gray")
```

```python
idxs = np.where(matched == np.amax(matched))
```

```python
idxs
```

```python
plt.imshow(flipped, cmap="gray")
```
